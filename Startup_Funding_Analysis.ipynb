```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv('startup_data.csv')

# Step 2: Clean and prepare data
# Ensure relevant columns exist
df = df[['Industry', 'Funding Amount (M USD)']].dropna()

# Step 3: Group by industry and calculate total + average funding
sector_trends = df.groupby('Industry')['Funding Amount (M USD)'].agg(['count', 'sum', 'mean']).sort_values(by='sum', ascending=False)

# Step 4: Display top 10 sectors by total funding
top_sectors = sector_trends.head(10)

# Step 5: Plot total funding by sector
plt.figure(figsize=(10, 6))
plt.barh(top_sectors.index[::-1], top_sectors['sum'][::-1], color='darkorange')
plt.xlabel('Total Funding (M USD)')
plt.title('Top 10 Sectors by Total Investment')
plt.tight_layout()
plt.show()

# Optional: Print summary table
print("Top 10 Sectors by Investment:\n")
print(top_sectors)
```
